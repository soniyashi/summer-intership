```python
import pygame
import sys

pygame.init()

WIDTH, HEIGHT = 300, 350
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Tic Tac Toe")

board = [
    ["", "", ""],
    ["", "", ""],
    ["", "", ""]
]

player = "X"
winner = None
game_over = False

font = pygame.font.Font(None, 80)
small_font = pygame.font.Font(None, 40)


def check_winner():
    # Rows
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] != "":
            return board[row][0]

    # Columns
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != "":
            return board[0][col]

    # Diagonals
    if board[0][0] == board[1][1] == board[2][2] != "":
        return board[0][0]

    if board[0][2] == board[1][1] == board[2][0] != "":
        return board[0][2]

    return None


def check_draw():
    for row in board:
        for cell in row:
            if cell == "":
                return False
    return True


running = True

while running:

    screen.fill((255, 255, 255))

    # Grid
    pygame.draw.line(screen, (0, 0, 0), (100, 0), (100, 300), 3)
    pygame.draw.line(screen, (0, 0, 0), (200, 0), (200, 300), 3)

    pygame.draw.line(screen, (0, 0, 0), (0, 100), (300, 100), 3)
    pygame.draw.line(screen, (0, 0, 0), (0, 200), (300, 200), 3)

    # Draw X and O
    for row in range(3):
        for col in range(3):
            text = font.render(board[row][col], True, (0, 0, 0))
            screen.blit(text, (col * 100 + 30, row * 100 + 15))

    # Winner message
    if winner:
        msg = small_font.render(f"{winner} Wins!", True, (255, 0, 0))
        screen.blit(msg, (90, 310))

    elif game_over:
        msg = small_font.render("Draw!", True, (0, 0, 255))
        screen.blit(msg, (120, 310))

    for event in pygame.event.get():

        if event.type == pygame.QUIT:
            running = False

        if (
            event.type == pygame.MOUSEBUTTONDOWN
            and not winner
            and not game_over
        ):

            x, y = event.pos

            if y < 300:

                row = y // 100
                col = x // 100

                if board[row][col] == "":

                    board[row][col] = player

                    winner = check_winner()

                    if winner is None:
                        if check_draw():
                            game_over = True
                        else:
                            player = "O" if player == "X" else "X"

    pygame.display.update()

pygame.quit()
sys.exit()
```
