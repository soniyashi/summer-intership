```python
PASSWORD = "redhat"
```

```python
def check_password():
    user_password = input("ENTER YOUR PASSWORD:")
    if user_password == PASSWORD:
        return True
```

```python
def write_note():
    if check_password():
        note= input("ENTER YOUR NOTE:")
        with open ("data.txt", "w") as file:
            file.write(note)
```

```python
def read_note():
    if check_password():
        file = open("data.txt","r")
        data = file.read()
        file.close
        print("\n ypur secret note is:")
        print(data)
```

```python
def append_note():
    if check_password():
        note = input("enter new note:")
        file.write("\n" + note)
        file.close()
        print("note added succesfully")
```

```python
def menu():
    while True:
        print("""
        =========secret note keeper=========
        1. write note
        2. read note
        3. append note
        4. exit
        """)
        choice = input("enter your choice :")
        if choice == "1":
            write_note()
        elif choice == "2":
            read_note()
        elif choice == "3":
            append_note()
        elif choice == "4":
            print("bye bye!!!!")
            break
        else:
            print("invalid choice, enter correct choice!!!!")
menu()
```
