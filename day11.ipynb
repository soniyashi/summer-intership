```python
class grandfather:
    def land(self):
        print("20 acre of lands ")
```

```python
class father (grandfather):
    def house (self):
        print("VIlla ")
```

```python
class child(father):
    def car(self):
        print("AUDI")
```

```python
s1 = child()
```

```python
s1.car()
```

```python
s1.house()
```

```python
s1.land()
```

```python
class parent:
    def display (self):
        print("I am parent Class")
```

```python
class child(parent):
    def display(self):
        super().display()
        print("I AM CHILD CLASS")
```

```python
ch = child()
```

```python
ch.display()
```

```python
class parent:
    def __init__(self):
        print("Parent Created")
```

```python
#GETTER METHOD
class bank:
    def __init__(self):
        self.__balance = 500
    def deposit(self , amount):
        self.__balance += amount
    def get_balance(self):
        return self.__balance
```

```python
b1 = bank()
```

```python
b1.deposit(300)
```
