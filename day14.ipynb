```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("data.csv")

# Input feature
x = df[["hr"]]

# Target variable
y = df["marks"]

# Create model
model = LinearRegression()
```

```python
# Train model
model.fit(x, y)
```

```python
model.coef_
```

```python
model.get_params()
```

```python
model.predict([[3]])
```

```python
plt.figure()
plt.scatter(df["hr"], df["marks"])
plt.plot(x , model.predict(x))
plt.xlabel("HRS")
plt.ylabel("marks")
plt.show()
```
