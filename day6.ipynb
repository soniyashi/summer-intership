```python
image = cv2.imread("panda.png")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

start_row , end_row = 100 , 200

start_col ,  end_col = 100 , 200

cropped = image_rgb[start_row:end_row , start_col:end_col]

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(cropped)
axes[1].set_title("Cropped Image")
axes[1].axis("off")

plt.show()
```

```python
image = cv2.imread("panda.png")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
height , width = image.shape[:2]
```

```python
center = (width//2 , height//2)
```

```python
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)
```

```python
rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].imshow(image_rgb)
axes[0].axis("off")

axes[1].imshow(rotated_rgb)
axes[1].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

canvas = np.ones((400, 600, 3), dtype="uint8") * 255

cv2.rectangle(canvas, (100, 100), (300, 250), (255, 0, 0), 3)

canvas_rgb = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

plt.imshow(canvas_rgb)
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

canvas = np.ones((400, 600, 3), dtype="uint8") * 255

cv2.circle(canvas, (300, 200), 80, (0, 255, 0), 3)

canvas_rgb = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

plt.imshow(canvas_rgb)
plt.axis("off")
plt.show()
```
