```python
import pygame
import random

# Initialize pygame
pygame.init()

# Screen size
WIDTH = 600
HEIGHT = 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Dodge the Block")

# Colors
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
BLACK = (0, 0, 0)

# Clock
clock = pygame.time.Clock()

# Player
player_size = 50
player_x = WIDTH // 2
player_y = HEIGHT - 70
player_speed = 7

# Block
block_size = 40
block_x = random.randint(0, WIDTH - block_size)
block_y = -40
block_speed = 5

# Score
score = 0
font = pygame.font.SysFont("Arial", 30)

running = True

while running:
    clock.tick(60)
    screen.fill(BLACK)

    # Events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Keyboard
    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT] and player_x > 0:
        player_x -= player_speed

    if keys[pygame.K_RIGHT] and player_x < WIDTH - player_size:
        player_x += player_speed

    # Move block
    block_y += block_speed

    # Reset block
    if block_y > HEIGHT:
        block_y = -40
        block_x = random.randint(0, WIDTH - block_size)
        score += 1
        block_speed += 0.2

    # Draw player and block
    pygame.draw.rect(screen, BLUE,
                     (player_x, player_y, player_size, player_size))
    pygame.draw.rect(screen, RED,
                     (block_x, block_y, block_size, block_size))

    # Collision
    player_rect = pygame.Rect(player_x, player_y,
                              player_size, player_size)
    block_rect = pygame.Rect(block_x, block_y,
                             block_size, block_size)

    if player_rect.colliderect(block_rect):
        game_over = font.render(
            f"GAME OVER! Score: {score}", True, WHITE)
        screen.blit(game_over, (120, HEIGHT // 2))
        pygame.display.update()
        pygame.time.delay(3000)
        running = False

    # Display score
    score_text = font.render(f"Score: {score}", True, WHITE)
    screen.blit(score_text, (20, 20))

    pygame.display.update()

pygame.quit()
```
